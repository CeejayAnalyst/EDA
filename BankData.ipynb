```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('/content/Cleaned_bank .csv')

df.head()
df.info()
df.describe()
```

```python
#Marital Status 

Marital_status = df['marital'].value_counts()
print(Marital_status)

Marital_status.plot(kind='bar')

plt.xlabel('Marital_status')
plt.ylabel('Count')
plt.title('Total count of clinets that are married or not')
```

Most of the clients in this dataset are married .

```python
#Exploring total number of Customers with a default in loan 

loan_defaults = df['Load_default'].value_counts()
print(loan_defaults)

loan_defaults.plot(kind='bar')

plt.xlabel('Loan default')
plt.ylabel('Count')
plt.title('number of loan default')
```

Most Customers in this dataset does not have loan defaults

```python
#Lets Explore Education column 

education_type = df['education'].value_counts()
print(education_type)

education_type.plot(kind='bar')

plt.xlabel('Loan default')
plt.ylabel('Count')
plt.title('number of loan default')
```

Most of the Customers in the dataset are University Graduates 

```python
#Lets examine the age column with a box plot

plt.hist(df['age'])

plt.xlabel('Age_group')
plt.ylabel('Age_count')
plt.title('Age distribution')

```

The dataset is made up of mostly customers of Ages 28 - 59

```python
#Exploring the loan , contact type , day_of _week column , housing loan

#Pie chart for percentage of the dataset that has loan
loan_counts = df['loan'].value_counts()
labels = loan_counts.index
sizes = loan_counts.values

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Percentage of Clients in Each Loan Category')
plt.show()


#Pie chart for percentage of the dataset that has loan
Housing_loan_counts = df['housing'].value_counts()
labels = Housing_loan_counts.index
sizes = Housing_loan_counts.values

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Percentage of Clients in Each Housing Loan Category')
plt.show()


#Barchart comparing different telemarketing contact method
contact_counts = df['contact'].value_counts()
labels = contact_counts.index
values = contact_counts.values

plt.bar(labels, values)
plt.xlabel('Contact Method')
plt.ylabel('Total Number of Clients')
plt.title('Total Number of Clients by Contact Method')
plt.show()

#Bar chart comparing the day of teh week clients were mostly contacted
day_of_week_counts = df['day_of_week'].value_counts()
labels = day_of_week_counts.index
values = day_of_week_counts.values

plt.bar(labels, values)
plt.xlabel('Day of Week')
plt.ylabel('Total Count of Clients Contacted')
plt.title('Total Count of Clients Contacted by Day of Week')
plt.show()
```

Shows that 53% of Customers in the dataset has housing loan , none has personal loans.
Most clients were contacted on Thursdays on their cellular line .

```python
#we will explore pdays . 
#which shows number of days that passed by after last contacted from a previous campaign 
#(numeric; 999 means client was not previously contacted)
```

```python

# Calculate the count of clients contacted and not contacted
contacted_count = len(df[df['pdays'] != 999])
print(contacted_count)
not_contacted_count = len(df[df['pdays'] == 999])

# Calculate the percentage of clients contacted and not contacted
total_count = len(df)
contacted_percentage = (contacted_count / total_count) * 100
not_contacted_percentage = (not_contacted_count / total_count) * 100

print("Count of Clients Contacted: ", contacted_count)
print("Count of Clients Not Contacted: ", not_contacted_count)
print("Percentage of Clients Contacted: ", contacted_percentage)
print("Percentage of Clients Not Contacted: ", not_contacted_percentage)
```
